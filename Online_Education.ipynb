```r
# Replace "your_file_path.csv" with the actual path to your CSV file
# If your CSV file is in the current working directory, you can just provide the file name directly.
# For example: read.csv("your_file.csv")
data <- read.csv("/content/ONLINE EDUCATION SYSTEM REVIEW.csv")
```

```r
head(data, n=5)
```

```r
# Load necessary packages (if not already loaded)
install.packages("ggplot2")  # Run this line if you haven't installed ggplot2 yet
library(ggplot2)

# Question 1: Distribution of students based on gender and level of education
# Create a bar plot
ggplot(data, aes(x = Gender, fill = Level.of.Education)) +
  geom_bar(position = "dodge") +
  labs(title = "Distribution of Students based on Gender and Level of Education",
       x = "Gender", y = "Count") +
  scale_fill_discrete(name = "Level of Education")
```

```r
# Load necessary libraries
library(dplyr)
library(ggplot2)

# Assuming you have already read the CSV file and stored it in a variable called 'data'

# Step 1: Filter data based on rural and urban locations
data_rural <- data %>% filter(Home.Location == "Rural")
data_urban <- data %>% filter(Home.Location == "Urban")

# Step 2: Calculate the average number of subjects for each group
avg_subjects_rural <- mean(data_rural$Number.of.Subjects)
avg_subjects_urban <- mean(data_urban$Number.of.Subjects)

# Step 3: Create a summary table for comparison
summary_table <- data.frame(
  Location = c("Rural", "Urban"),
  Average_Subjects = c(avg_subjects_rural, avg_subjects_urban)
)

# Step 4: Print the summary table
print(summary_table)

# Step 5: Create a bar plot to visualize the comparison
ggplot(summary_table, aes(x = Location, y = Average_Subjects, fill = Location)) +
  geom_bar(stat = "identity", position = "dodge") +
  labs(title = "Average Number of Subjects Taken by Students\nBased on Home Location",
       x = "Home Location",
       y = "Average Number of Subjects")
```

```r
# Bar plot using ggplot2
ggplot(data, aes(x = Device.type.used.to.attend.classes, y = Time.spent.on.social.media..Hours.)) +
  geom_bar(stat = "summary", fun = "mean", fill = "steelblue") +
  labs(title = "Time Spent on Social Media by Device Type",
       x = "Device Type",
       y = "Average Time Spent on Social Media (Hours)")
```

```r
# Calculate the average study time for students interested and not interested in gaming
average_study_time <- aggregate(Study.time..Hours. ~ Interested.in.Gaming., data = data, FUN = mean)
# Print the result
print(average_study_time)
# Create a bar plot to visualize the average study time for students interested and not interested in gaming
ggplot(average_study_time, aes(x = Interested.in.Gaming., y = Study.time..Hours.)) +
  geom_bar(stat = "identity", fill = "steelblue") +
  labs(title = "Average Study Time by Interest in Gaming",
       x = "Interested in Gaming",
       y = "Average Study Time (Hours)")
```
